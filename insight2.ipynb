I'm going to start everything off by just jumping right into a data set that can be found at https://catalog.data.gov/dataset/table-iii-deaths-in-122-u-s-cities. I've downloaded the CSV file, renamed it to 'deaths.csv', and placed it in a directory called 'data' inside my working folder. 

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Here we've just loaded some of the things we'll need for what we're going to do with this data. Using the % symbol is what's appropriately called 'magic' in iPython, and in this case it will allow us to display plots right inside our notebook. After we've loaded in some of the necessities, we'll go ahead and read the data into a pandas dataframe:

```python
deaths = pd.read_csv('./data/deaths.csv')
```

I happen to know that this is a Comma Separated Values file (note: .csv). If we didn't already know that, there are ways of figuring out how to read data. We'll take a look at that at a later date. For now, let's direct our attention to the first 5 indices of the dataframe we just made:

```python
deaths[:5]
```

I'm not really sure what information is held in the columns ending in 'flag', at at this point I don't really care. What I'm going to concern myself with first is the number of deaths in the different areas across the given categories. So let's go ahead and tidy up the table a little bit, getting rid of the columns that I'm not going to use:

```python
deaths_no_flag = deaths.drop([col for col in deaths.columns if 'flag' in col], axis = 1)
```

Note that we're dropping from axix = 1, this is specifying that we're dropping columns and not rows (that would be axis = 0). So let's see what the first few entries look like in our new table:

```python
deaths_no_flag[:5]
```

But don't worry, the original dataframe, deaths, is still available for us if we want to use it. So let's get some information from it:

```python
deaths.shape
```

We can see that the deaths dataframe has 6996 rows, this will be the same number of rows as in the deaths_no_flag dataframe since we only dropped some columns. Since we're interested in these deaths with relation to where they were reported, let's see how many times each unique 'Reporting Area' shows up:

```python
deaths_no_flag['Reporting Area'].value_counts()
```

Hmmm.. Each shows up exactly 53 times. Well, if we had been really smart, we could have read the description in the url I started this with and realized that this is probably one entry per week per Reporting Area. But, meh, why spoil the fun? Let's investigate this by splitting the dataframe into multiple smaller dataframes, stored in a dictionary, each of which only contains one Reporting Area:

```python
areas = {}
unique_areas = pd.unique(deaths_no_flag['Reporting Area'].ravel())
for area in unique_areas:
    areas[area] = deaths_no_flag[deaths_no_flag['Reporting Area'] == area]
```

Before we investigate further, we should probably take a look at what unique areas we actually have to deal with, because these are the keys in our new areas dictionary.

```python
unique_areas
```

Because New England is first on the list (and because I won't have to type a comma in every time I want to check it out), let's see what its dataframe looks like:

```python
areas['New England']
```

The first thing that I notice looking at this, is that our index (inherited from the deaths dataframe) is no longer very useful. The second thing I notice is that we were right about there being one entry per week in each region, or at least in New England and I'll trust (from the data discription via the source) that it's true for the rest as well. This is great! We have a natural index handed to us in the form of the 'MMWR WEEK' column. 'MMWR' stands for Morbidity and Mortality Weekly Report, by the way, this is where our data is coming from. So, after that most pleasant of acronymns, let's go ahead and make that our index over all of the regions:

```python
for area in unique_areas:
    areas[area] = areas[area].set_index(areas[area]['MMWR WEEK'])
```

We should now have a slightly nicer looking dataframe for New England (and in fact, for all of them):

```python
areas['New England']
```

Because we have the MMWR WEEK column as our index (and because pandas is so good to us and labels it) we can go ahead and drop the MMWR WEEK column as it is now superflous. In fact, we can kill two birds with one stone and remove the MMWR YEAR column as well by dropping any column that has the string 'MMWR' in its title. The reason I suggest dropping the year column is because all of our data pertains to 2014. If we so wished we could use the value_counts() method on deaths['MMWR YEAR'] to verify this.

```python
for area in unique_areas:
    areas[area] = areas[area].drop([col for col in areas[area].columns if 'MMWR' in col], axis = 1)
```

And now for the lovely tidied up table for New England, remember that we could replace 'New England' with any of the other keys in unique_areas to get the corresponding table for that area.

```python
areas['New England']
```
